```julia
using JLD
using BenchmarkTools
using PyPlot
```

```julia
var500 = load("./var500.jld")
d1 = var500["d1"]
r1 = var500["r1"]
r2 = var500["r2"]
r3 = var500["r3"]
r4 = var500["r4"]
r7 = var500["r7"]
de1 = var500["de1"]
de2 = var500["de2"]
pr = var500["pr"]
```

```julia
length(d1)
```

```julia
dt = d1.times[2:18]/1000
rt = (r1.times[1:13] + r2.times[1:13] + r3.times[1:13] + r4.times[1:13] + r7.times[1:13])/1000
det = (de1.times[2:18] + de2.times[2:18] + pr.times[2:18])/1000
```

```julia
dt = d1.times/1000
rt = (r1.times[2:13] + r2.times[2:13] + r3.times[2:13] + r4.times[2:13] + r7.times[2:13])/1000
det = (de1.times + de2.times + pr.times)/1000
```

```julia
fig = figure("pyplot_boxplot",figsize=(19,5))
subplot(141)
boxplot([dt,rt,det])
title("Default, Repayment and Decision")
subplot(142)
boxplot(dt)
title("Default")
subplot(143)
boxplot(rt)
title("Repayment")
subplot(144)
boxplot(det)
title("Decision")

suptitle("Box Plot, Run Time of Default, Repayment and Decision section in (μs)")
```
